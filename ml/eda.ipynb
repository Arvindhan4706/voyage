# Exploratory Data Analysis: Flight Prices
This notebook explores our synthetic flight pricing dataset to satisfy the requirements for Review I.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/flight_prices.csv')
df.head()
```

## 1. Distribution of Flight Prices

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price_inr'], bins=50, kde=True)
plt.title('Distribution of Flight Prices (INR)')
plt.xlabel('Price (INR)')
plt.ylabel('Frequency')
plt.show()
```

## 2. Price vs. Distance

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='distance_km', y='price_inr', data=df, alpha=0.3)
plt.title('Flight Price vs Distance')
plt.xlabel('Distance (km)')
plt.ylabel('Price (INR)')
plt.show()
```

## 3. Impact of Booking Advance (Days to Departure)

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='days_to_departure', y='price_inr', data=df)
plt.title('Average Flight Price by Days to Departure')
plt.xlabel('Days to Departure')
plt.ylabel('Average Price (INR)')
plt.xlim(0, 60)
plt.show()
```
